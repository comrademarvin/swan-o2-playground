This is not really what I want yet, but it's a first step. This notebook runs in plain ROOT, opens a `o2sim_HitsTRD.root` file (that was previously generated with `o2-sim -n 5`), accesses the o2sim tree and plots some positions.

## Open the ROOT file

Here we open the root file and connect our tree. The dictionaries, e.g. for `o2::trd::Hit` are not available, but the loading still works.

```c++
TFile f("/eos/home-t/tdietel/tmp/sim/o2sim_HitsTRD.root");
f.ls();
TTree* o2sim;
f.GetObject("o2sim", o2sim);
```

## Plot something

```c++
TCanvas cnv;
o2sim->Draw("locC:locR");
cnv.Draw();
```
